# Introduction to Pneumonia

Pneumonia is a serious lung infection caused by various pathogens, including bacteria, viruses, or fungi. It triggers inflammation within the alveoli, small air sacs essential for gas exchange, resulting in the accumulation of fluid or pus and compromising respiratory function.<br>
The primary mode of transmission involves airborne droplets released during coughing or sneezing, with additional risk posed by contact with contaminated surfaces.

Diagnostic assessment typically involves chest X-ray imaging, allowing healthcare providers to precisely determine the type and severity of pneumonia. This condition is categorized into three primary types:

- **Bacterial pneumonia**, often attributed to pathogens like Streptococcus pneumoniae or Haemophilus influenzae.
- **Viral pneumonia**, frequently caused by influenza or respiratory syncytial virus (RSV).
- **Fungal pneumonia**, predominantly affecting individuals with weakened immune systems.

Each category presents unique causal factors and clinical manifestations, necessitating tailored treatment and management approaches.

```python
# Define paths
train_path = "./chest_xray/train/"
test_path = "./chest_xray/test/"
val_path = "./chest_xray/val/"

def process_dataset(directory):
    image_paths = []
    labels = []
    for label in os.listdir(directory):
        label_path = os.path.join(directory, label)
        for image in os.listdir(label_path):
            image_paths.append(os.path.join(label, image))  # Store relative paths
            labels.append(label)
    return pd.DataFrame({'image_paths': image_paths, 'labels': labels})

# Process datasets
train_df = process_dataset(train_path)
test_df = process_dataset(test_path)
val_df = process_dataset(val_path)
```

```python
import os
import matplotlib.pyplot as plt

# Count images in each class for each dataset
train_pneumonia_count = len(os.listdir(train_path + "/PNEUMONIA"))
train_normal_count = len(os.listdir(train_path + "/NORMAL"))

test_pneumonia_count = len(os.listdir(test_path + "/PNEUMONIA"))
test_normal_count = len(os.listdir(test_path + "/NORMAL"))

val_pneumonia_count = len(os.listdir(val_path + "/PNEUMONIA"))
val_normal_count = len(os.listdir(val_path + "/NORMAL"))

train_data = [train_pneumonia_count, train_normal_count]
test_data = [test_pneumonia_count, test_normal_count]
val_data = [val_pneumonia_count, val_normal_count]

# Labels 
labels = ['PNEUMONIA', 'NORMAL']

# Create subplots with smaller figure size
fig, axes = plt.subplots(3, 2, figsize=(7, 7))

# Colors
colors_pie = ['#505050', '#101010']  
colors_bar = ['#505050', '#101010']  

# Training data set charts
axes[0, 0].pie(train_data, labels=labels, autopct='%1.1f%%', colors=colors_pie, textprops={'color': 'white'})
axes[0, 0].set_title('Training data set - Pie Chart')

axes[0, 1].bar(labels, train_data, color=colors_bar)
axes[0, 1].set_title('Training data set - Bar Chart')

# Test data set charts
axes[1, 0].pie(test_data, labels=labels, autopct='%1.1f%%', colors=colors_pie, textprops={'color': 'white'})
axes[1, 0].set_title('Test data set - Pie Chart')

axes[1, 1].bar(labels, test_data, color=colors_bar)
axes[1, 1].set_title('Test data set - Bar Chart')

# Validation data set charts
axes[2, 0].pie(val_data, labels=labels, autopct='%1.1f%%', colors=colors_pie, textprops={'color': 'white'})
axes[2, 0].set_title('Validation data set - Pie Chart')

axes[2, 1].bar(labels, val_data, color=colors_bar)
axes[2, 1].set_title('Validation data set - Bar Chart')

plt.tight_layout()
plt.show()
```

```python
# Function to display images
def display_images(data_dir, labels, num_images=1):
    plt.figure(figsize=(20, 20))
    
    image_count = 1
    for label in labels:
        path = os.path.join(data_dir, label)
        images = os.listdir(path)[:num_images]  # Select only the first 'num_images' images
        
        for image in images:
            plt.subplot(3, 4, image_count)
            img = plt.imread(os.path.join(path, image))
            plt.imshow(img, cmap='gray')
            plt.axis("off")
            plt.title(f"{label} X-ray")
            image_count += 1
            
    plt.tight_layout()
    plt.show()
```

```python
# Display images from each dataset
print("Training set:")
display_images(train_path, labels)

print("Testing set:")
display_images(test_path, labels)

print("Validation set:")
display_images(val_path, labels)
```

```python
# Define batch size and image size
batch_size = 16
image_size = (128, 128)

# Function to create data generators
def create_datagen(df, directory, shuffle, is_training):
    if is_training:
        datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=40,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.3,
            horizontal_flip=True,
            fill_mode='nearest'
        )
    else:
        datagen = ImageDataGenerator(rescale=1./255)
    
    return datagen.flow_from_dataframe(
        dataframe=df,
        directory=directory,
        x_col="image_paths",  # Use the correct column name
        y_col="labels",       # Use the correct column name
        target_size=image_size,
        batch_size=batch_size,
        class_mode="binary",  # For binary classification
        color_mode="rgb",
        shuffle=shuffle,  # Shuffle only for training data
        seed=42       # Set seed for reproducibility
    )

# Create data generators for training, validation, and testing datasets
train_generator = create_datagen(train_df, train_path, shuffle=True, is_training=True)
valid_generator = create_datagen(val_df, val_path, shuffle=True, is_training=False)
test_generator = create_datagen(test_df, test_path, shuffle=False, is_training=False)

# Display the class indices to verify correctness
print("Class indices:", train_generator.class_indices)
```
